## Sinusoid Data Generation
First we create the dataset $X = \left\lbrace x_i\right\rbrace_{i = 1}^{n_{\Phi}}$ of $n_\Phi$ datapoints. Each data point $x_i$ is given in terms of a sinusoidal function.

\begin{equation}
x_i(t) = \text{sin}(\omega t +\varphi_i)
\end{equation}

Each data point has $n_T$ dimensions in terms of the samples taken from the regular partition of the $(0,1]$ interval $t\in \left\lbrace\frac{1}{n_T}, \frac{2}{n_T}, \frac{3}{n_T}, \ldots, \frac{(n_T-1)}{n_T}, 1\right\rbrace$.The phase $\varphi$ defines each of the datapoints and is obtained from the regular partition of the interval $(0,2\pi]$ $\varphi \in \left\lbrace\frac{2\pi}{n_{\Phi}}, \frac{4\pi}{n_{\Phi}},\ldots, \frac{2\pi(n_{\Phi}-1)}{n_\Phi}, 2\pi\right\rbrace$. The angular frequency $\omega$ in this experiment is fixed.
   

```python
def sinusoid_bundle_generation(n_Phi, n_T, omega):
    phase_range = 2*np.pi*np.linspace(0,1,n_Phi)
    time_range = np.linspace(0,1,n_T)
    signals = np.sin(np.subtract.outer(phase_range,-(omega * time_range)))
    # Signals
    return phase_range, time_range, signals

n_Phi = 100
n_T = 50
omega = 2*np.pi
phase_range, time_range, signals = sinusoid_bundle_generation(n_Phi, n_T, omega)
signals = signals[:,:,np.newaxis]
```

## Convolutional VAE

The VAE used in this experiment makes use of 1D convolutions and convolutional transpose (commonly called also deconvolution, which is an incorrect term). Each of the data points has a dimension of $n_T$   

```python
# Parameters
input_dim = n_T
latent_dim = 2 # 
intermediate_dim = int(input_dim/3) # Intermediate dense layer size
epochs = 400 # Number of training epochs
batch_size = 100
filters = 64 # Number of filters obtained in convolution
kernel_size = 3 # Kernel size of the deconvolutions
epsilon_std = 1.0 #Standard deviation for sampling epsilon  
```

# Convolutional VAE

```python
# Deconvolutional layers of 1D are still not defined in keras.
def Conv1DTranspose(input_tensor, filters, kernel_size, strides=2, padding='same', activation='relu'):
    x = Lambda(lambda x: K.expand_dims(x, axis=2))(input_tensor)
    x = Conv2DTranspose(filters=filters, kernel_size=(kernel_size, 1), strides=(strides, 1), padding=padding, activation=activation)(x)
    x = Lambda(lambda x: K.squeeze(x, axis=2))(x)
    return x
```

```python
# ENCODING
x = Input(shape=(input_dim,1), name='x')
# Conv 1D Layers
c = Conv1D(32,3,input_shape = (input_dim,1), strides = 1, padding='valid', activation = 'relu', name = 'c')(x)
c2 = Conv1D(32,3, strides = 1, padding='valid', activation = 'relu', name = 'c2')(c)
# Dense Layers
c_flatten = Flatten()(c2)
h = Dense(intermediate_dim, activation='relu', name='h')(c_flatten)
z_mean = Dense(latent_dim, name='z_mean')(h)
z_log_sigma = Dense(latent_dim, name='z_log_sigma')(h)
# Sampling from the encoder distribution q(z|x)
def sampling(args):
    z_mean, z_log_sigma = args
    epsilon = K.random_normal(shape=(batch_size, latent_dim),
                              mean=0., stddev=epsilon_std)
    return z_mean + K.exp(z_log_sigma) * epsilon
z = Lambda(sampling, output_shape=(latent_dim,), name='sampling')([z_mean, z_log_sigma])




# DECODING
# First we define the layers which are used in the VAE and in the decoder network
# These layers are shared between the VAE and the decoder 
decoder_upsample = Dense(filters*int(input_dim//2), activation = 'relu', name = 'upsampling')
decoder_reshape = Reshape((int(input_dim//2),filters))
decoder_deconv1 = Conv1D(filters, kernel_size,strides=1, padding = 'same', activation='relu')
decoder_deconv2 = Conv1D(filters, kernel_size,strides=1, padding = 'same', activation='relu')
decoder_mean_squash = Conv1D(1, 2,strides = 1, padding='valid', activation='linear', name = 'output')
flatten_dimension = Lambda(lambda x: K.squeeze(x, axis=2))

# VAE decoding
z_upsample = decoder_upsample(z)
z_reshape = decoder_reshape(z_upsample)
z_deconv1 = decoder_deconv1(z_reshape)
z_deconv2 = decoder_deconv2(z_deconv1)
z_deconv3_upsample = Conv1DTranspose(z_deconv2, filters, kernel_size, strides = 2, padding = 'valid')
z_mean_squash = decoder_mean_squash(z_deconv3_upsample)
vae = Model(x,z_mean_squash)

# Decoder
_z = Input(shape=(latent_dim,), name='basic_decoder_input')
_z_upsample = decoder_upsample(_z)
_z_reshape = decoder_reshape(_z_upsample)
_z_deconv1 = decoder_deconv1(_z_reshape)
_z_deconv2 = decoder_deconv2(_z_deconv1)
_z_deconv3_upsample = Conv1DTranspose(_z_deconv2, filters, kernel_size, strides = 2, padding = 'valid')
_z_mean_squash = decoder_mean_squash(_z_deconv3_upsample)

decoder = Model(_z,_z_mean_squash)
encoder = Model(x, z_mean)
encoder_sigma = Model(x, z_log_sigma)

# LOSS FUNCTIONS
def xent_loss(x,x_decoded_mean):
    xent_loss = K.mean(losses.mean_squared_error(x,x_decoded_mean))
    return xent_loss
def kl_loss(x, x_decoded_mean):
    kl_loss = -0.5 * K.mean(1 + z_log_sigma - K.square(z_mean) - K.exp(z_log_sigma), axis=-1)
    return kl_loss
def vae_loss(x, x_decoded_mean):
    return xent_loss(x,x_decoded_mean) + kl_loss(x,x_decoded_mean)
```

## Callbacks

- **Reduce Learning Rate on Plateau**: This callback multiplies the learning rate by a certain factor if the monitored quantity has not decreased after a certain number of steps defined by its patience.
- **Tensorboard**: This callback allows to monitor scalar quantities. 
- **Model Checkpoint**: Allows to save the best model that has the lowest loss for the validation set. I have not really used this. 

```python
# File naming variables
experiment_name = 'sinusoid_1Dlatent_UP' # Code for the experiment
timestr = time.strftime("%Y-%m-%d-%H-%M_") # Timestamp
extra_comments ='_' # Extra comments of the experiment




# Folder for saving the different models
folder_models = join(getcwd(),'models', experiment_name)
if not exists(folder_models):
    makedirs(folder_models)
    
# Folder for the generator, the encoder and the vae models. Also tensorboard
folder_dictionary = {'generator':join(folder_models,'generator'),
                    'encoder':join(folder_models,'encoder'),
                    'encoder_sigma':join(folder_models,'encoder_sigma'),
                    'vae':join(folder_models,'vae'),
                    'tensorboard':join(folder_models,'tensorboard'),
                    'images':join(folder_models,'images')}
for folder in folder_dictionary:
    if not exists(folder_dictionary[folder]):
        makedirs(folder_dictionary[folder])
        
log_dir_tensorboard =join(win32api.GetShortPathName(folder_dictionary['tensorboard']),timestr, extra_comments)
```

```python
reduce_lr = ReduceLROnPlateau(monitor = 'xent_loss',
                              factor = 0.5,
                              patience = 20,
                              mode = 'min',
                              epsilon = 0.000001)

tbCallback = TensorBoard(log_dir =log_dir_tensorboard)

model_checkpoint = ModelCheckpoint(filepath = join(folder_dictionary['vae'],'BEST'),
                                  monitor = 'val_loss',
                                  save_best_only = True,
                                  save_weights_only = True)

# vae.fit_generator(data_generator,
#                   epochs=epochs,
#                   steps_per_epoch=num_batches,
#                   callbacks=[reduce_lr, tbCallback],
#                   validation_data=None, shuffle = False, workers = 1, max_queue_size = 1)

vae.fit(signals,signals,
                  epochs=epochs,
                    batch_size=100,
                  callbacks=[ tbCallback],
                  validation_data=None, shuffle = False)

# Save the trained models 
experiment_instance = timestr+extra_comments
vae.save_weights(join(folder_models, experiment_instance+'.h5'))
generator.save(join(folder_dictionary['generator'], experiment_instance))
encoder.save(join(folder_dictionary['encoder'], experiment_instance))
encoder_sigma.save(join(folder_dictionary['encoder_sigma'], experiment_instance))
vae.save(join(folder_dictionary['vae'], experiment_instance))
```

## Evaluation of the VAE

From the VAE we would like to evaluate the reconstruction capabilities of the network. A signal is given to the network to reconstruct it by the VAE. The output is compared visually with the input signal for each of the different phases from the partition. 

```python
predictions = vae.predict(signals, batch_size = batch_size)
```

```python
%matplotlib inline
fig, ax = plt.subplots(10,10)
fig.set_figheight(20)
fig.set_figwidth(20)
for num_row, row in enumerate(ax):
    for num_col, col in enumerate(row):
        col.plot(time_range, predictions[np.ravel_multi_index((num_row,num_col),(10,10))],'-x')
        col.plot(time_range, signals[np.ravel_multi_index((num_row,num_col),(10,10))],'-x')
        col.set_xticks([])
        col.set_yticks([])
        col.set_xlabel(str(int(360*phase_range[np.ravel_multi_index((num_row,num_col),(10,10))]/(2*np.pi))))
        col.grid()
plt.title('Input-Output signal comparison from VAE')
plt.show()
plt.savefig('Conv2DLatent2')
```

```python
import matplotlib.pyplot as plt
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
import numpy as np; np.random.seed(42)
import PyQt5
%matplotlib qt5
# Generate data x, y for scatter and an array of images.
x = np.arange(20)
y = np.random.rand(len(x))
arr = np.empty((len(x),10,10))
for i in range(len(x)):
    f = np.random.rand(5,5)
    arr[i, 0:5,0:5] = f
    arr[i, 5:,0:5] =np.flipud(f)
    arr[i, 5:,5:] =np.fliplr(np.flipud(f))
    arr[i, 0:5:,5:] = np.fliplr(f)

# create figure and plot scatter
fig = plt.figure()
ax = fig.add_subplot(111)
line, = ax.plot(x,y, ls="", marker="o")

# create the annotations box
im = OffsetImage(arr[0,:,:], zoom=5)
xybox=(50., 50.)
ab = AnnotationBbox(im, (0,0), xybox=xybox, xycoords='data',
        boxcoords="offset points",  pad=0.3,  arrowprops=dict(arrowstyle="->"))
# add it to the axes and make it invisible
ax.add_artist(ab)
ab.set_visible(False)

def hover(event):
    # if the mouse is over the scatter points
    if line.contains(event)[0]:
        # find out the index within the array from the event
        ind, = line.contains(event)[1]["ind"]
        # get the figure size
        w,h = fig.get_size_inches()*fig.dpi
        ws = (event.x > w/2.)*-1 + (event.x <= w/2.) 
        hs = (event.y > h/2.)*-1 + (event.y <= h/2.)
        # if event occurs in the top or right quadrant of the figure,
        # change the annotation box position relative to mouse.
        ab.xybox = (xybox[0]*ws, xybox[1]*hs)
        # make annotation box visible
        ab.set_visible(True)
        # place it at the position of the hovered scatter point
        ab.xy =(x[ind], y[ind])
        # set the image corresponding to that point
        im.set_data(arr[ind,:,:])
    else:
        #if the mouse is not over a scatter point
        ab.set_visible(False)
    fig.canvas.draw_idle()

# add callback for mouse moves
fig.canvas.mpl_connect('motion_notify_event', hover)           
plt.show()
```
